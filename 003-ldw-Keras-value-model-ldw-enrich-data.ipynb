```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, classification_report
import keras
from keras.models import Sequential
from keras.layers import *
import tensorflow as tf
from keras import backend as K
pd.options.display.max_columns = 200

%matplotlib inline
```

### 1.0 Load training and test data

```python
%%time
# Building Permit Data
train = pd.read_csv('../data/working/DE_transformed.csv', index_col=0)
# test = pd.read_csv('../data/working/', index_col=0)

print(df.shape)
```

### 2.0 Augmented features

```python
# x1**2, x2**2, x1x2, sin(x1), sin(x2)
train['Lat_sq'] = train['Latitude']**2
train['Long_sq'] = train['Longitude']**2
train['Lat_x_Long'] = train['Longitude'] * train['Latitude']
train['sine_lat'] = np.sin(train['Latitude'])
train['sine_long'] = np.sin(train['Longitude'])

# ---- Test Augmentation ---- 
'''
# test['Lat_sq'] = test['Latitude']**2
# test['Long_sq'] = test['Longitude']**2
# test['Lat_x_Long'] = test['Longitude'] * test['Latitude']
# test['sine_lat'] = np.sin(test['Latitude'])
# test['sine_long'] = np.sin(test['Longitude'])
''''''
```

```python
# Train test split
X_train = train.drop('Value', axis=1)
y_train = train['Value']

# X_test = test.drop('logValue', axis=1)
# y_test = test['logValue']

print('Training shape:', X_train.shape, y_train.shape)
# print(' Testing shape:', X_test.shape, y_test.shape)
```

```python
# ================= Start Keras Session ==================== #

K.clear_session()
model = Sequential()

model.add(Dense(60, input_dim=X_train.shape[1], name='layer_1'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(Activation('relu'))

model.add(Dense(40, name='layer_2'))
model.add(Dropout(0.1))
model.add(BatchNormalization())
model.add(Activation('relu'))

model.add(Dense(20, activation='relu', name='layer_3'))
model.add(Dense(20, activation='relu', name='layer_4'))
model.add(Dense(20, activation='relu', name='layer_5'))

model.add(Dense(1, activation='linear', name='output_layer'))
model.compile(loss='mean_squared_error', optimizer='adam')

RUN_NAME = '26 MASSIVE Augmentation D(60d0.2-40d0.1-20x3), b_s 100, e500'
logger = keras.callbacks.TensorBoard(
            log_dir='logs/{}'.format(RUN_NAME),
            histogram_freq=0,
            write_graph=True)

# ---------------- Train the Model -------------------- #

model.fit(X_train.values, np.log(y_train.values),
          batch_size=100,
          epochs=500,
          shuffle=True,
          verbose=1,
          callbacks=[logger],
          validation_split=0.2)

# ------------------------------------------------------ #
```
